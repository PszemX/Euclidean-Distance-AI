# Tensorflow neural network AI

```python
import tensorflow as tf
import numpy as np
import math
import generators
from sklearn.metrics import accuracy_score
```

Training data

```python
gen = generators.GeneratorI(80)
x_train = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_train = np.array([[0], [1], [1], [2]])

x_test = np.array([[24, 1], [3, 1], [2, 55]])
y_test = np.array([[25], [4], [57]])
```

Create architecture

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=4, activation='relu'))
model.add(tf.keras.layers.Dense(units=2, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='relu'))
```

Compile and train

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mse')
# Train
model.fit(x_train, y_train, epochs=300)
```

Check result

```python
y_hat = model.predict(x_test)
y_hat = [ math.floor(val) if val < math.floor(val) + 0.5 else math.ceil(val) for val in y_hat ]
print(f"Accuracy score : {accuracy_score(y_test, y_hat)}")
print(model.predict(x_test)) 
```
